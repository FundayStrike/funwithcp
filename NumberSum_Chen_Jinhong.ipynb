Inspired by another The Devil's Plan game, Number Sum (https://www.youtube.com/watch?v=SNuedzRh2O4), write Python code to generate a game for a player to form numbers summing up to a target integer k. A game flow is as follows:
- display a 2D m by n grid of numbers. each number is between 1 and 9 inclusive.
- when the player is ready (by pressing y), computer will replace the grid, 'hiding' the numbers with ascending alphabets A, B, C, ...
- computer will generate a pseudo random up to 2-digit integer k
- prompt the player to enter a sequence of i letters which associated numbers sum up to k. the sequence must be connected either horizontally, vertically or diagonally. for a start, you may assume i = 3.
- if the sequence correctly sums up to k, player earns 1 point
- if the sequence does not sum up to k, player loses 1 point
- there can be more than 1 correct sequence that sums up to k. so prompt for input again.
- game terminates when
  - player has entered all sequences that sum to k
  - it is impossible to form a sequence that sums to k
  - player quits by entering -1  
- when the game terminates, show all possible sequences that sum to k, or 0 if none.

You may state reasonable assumptions for a minimum viable product. Have fun. :)

```python
import random
import string
import itertools

def oneDArray(x):
    return list(itertools.chain(*x))

#assuming the grids is 5x5
numgrid = []
for i in range(5):
    tmp = []
    for j in range(5):
        tmp.append(random.randint(1,9))

    numgrid.append(tmp)
    

#getting a 1d list to make it easier to check
oneDlist = oneDArray(numgrid)
    


#displaying number grid
for num in numgrid:
    print(num)
print('')
print('Remember the positions.')
print('')
#prints multiple blanks in order to push the numbergrid up so that user cant see it
def clearpg():
    for i in range(100):
        print('')
    print("Don't scroll up to cheat!!!")


#The number to get. Max is only 27 as i am assuming the number of letters to key in to be 3
k = random.randint(10,27)

x = input("Enter 'y' when you are ready: ",)
print('')
print('')
if x == 'y':
    clearpg()

#To get split the alphabets 5 by 5
alphabet = list(string.ascii_uppercase)
#temporary list to output the rows of alphabet
lst = alphabet[:len(oneDlist)]
output = []
while True:
    if len(lst) <= 0:
        break

    output.append(lst[:5])
    del lst[:5]


#To print the covered list:
for i in output:
    print(i)


print('')
print('')
print('The number should add up to ',k)
print('You should enter 3 letters.')
print('The letters should be in either horizontally, vertically or diagonally')
letter_index = []
for i in range(3):
    a = input('Enter letter '+ str(i+1)+': ',).upper()
    #convert it to index so we can check the sequence
    letter_index.append(alphabet.index(a))
print(letter_index)
go_on = True
for index in range(1,len(letter_index)):
    #the check values are these as if its beside each other, it will be one, if under it will be 5 if diagonal, it will be 6
    
    check = [1,5,6]
    diff = letter_index[index] - letter_index[index-1]
    if diff not in check:
        go_on = False
        print('You did not enter in sequence :(')
        break
if go_on:
    total = 0
    #The have a corressponding index
    for item in letter_index:
        total += oneDlist[item]

if total == k:
    print('You are CORRECT!!!')
else:
    print('You are wrong :(')






        

    


    

    
    





    
    
    
    

```
